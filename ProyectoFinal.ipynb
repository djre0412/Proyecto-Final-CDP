##### **Lectura de tablas**

```python
sql_query = 'SELECT *  FROM flights;'
table_flights= pd.read_sql(sql_query, mysql_driver)
suffix='_flights'
new_columns = [col + suffix for col in table_flights.columns]
table_flights.rename(columns=dict(zip(table_flights.columns, new_columns)), inplace=True)


sql_query = 'SELECT *  FROM airlines;'
table_airlines= pd.read_sql(sql_query, mysql_driver)
suffix='_airlines'
new_columns = [col + suffix for col in table_airlines.columns]
table_airlines.rename(columns=dict(zip(table_airlines.columns, new_columns)), inplace=True)

sql_query = 'SELECT *  FROM airport;'
table_airport= pd.read_sql(sql_query, mysql_driver)
suffix='_airport'
new_columns = [col + suffix for col in table_airport.columns]
table_airport.rename(columns=dict(zip(table_airport.columns, new_columns)), inplace=True)

sql_query = 'SELECT *  FROM planes;'
table_planes= pd.read_sql(sql_query, mysql_driver)
suffix='_planes'
new_columns = [col + suffix for col in table_planes.columns]
table_planes.rename(columns=dict(zip(table_planes.columns, new_columns)), inplace=True)

sql_query = 'SELECT *  FROM weather;'
table_weather= pd.read_sql(sql_query, mysql_driver)
suffix='_weather'
new_columns = [col + suffix for col in table_weather.columns]
table_weather.rename(columns=dict(zip(table_weather.columns, new_columns)), inplace=True)

```

##### **Cubo de datos**

```python
cubo=table_flights.merge(table_airlines, left_on='carrier_flights',right_on='carrier_airlines',how='left')
cubo=cubo.merge(table_airport, left_on='origin_flights',right_on='faa_airport',how='left')
cubo=cubo.merge(table_planes, left_on='tailnum_flights', right_on='tailnum_planes' ,how='left')
cubo=cubo.merge(table_weather, left_on=['origin_flights','year_flights','month_flights','day_flights','hour_flights'], right_on=['origin_weather','year_weather','month_weather','day_weather','hour_weather'] ,how='left')


cubo['dep_time_flights']=cubo['dep_time_flights'].astype(float)
cubo['sched_dep_time_flights']=cubo['sched_dep_time_flights'].astype(int)
cubo['dep_delay_flights']=cubo['dep_delay_flights'].astype(float)
cubo['arr_time_flights']=cubo['arr_time_flights'].astype(float)
cubo['sched_arr_time_flights']=cubo['sched_arr_time_flights'].astype(int)
cubo['arr_delay_flights']=cubo['arr_delay_flights'].astype(float)
cubo['air_time_flights']=cubo['air_time_flights'].astype(float)
cubo['distance_flights']=cubo['distance_flights'].astype(int)
cubo['distance_flights']=cubo['distance_flights'].astype(int)
cubo['time_hour_flights'] = pd.to_datetime(cubo['time_hour_flights'], errors='coerce')
cubo['temp_weather']=cubo['temp_weather'].astype(float)
cubo['dewp_weather']=cubo['dewp_weather'].astype(float)
cubo['humid_weather']=cubo['humid_weather'].astype(float)
cubo['wind_dir_weather']=cubo['wind_dir_weather'].astype(float)
cubo['wind_speed_weather']=cubo['wind_speed_weather'].astype(float)
cubo['wind_gust_weather']=cubo['wind_gust_weather'].astype(float)
cubo['pressure_weather']=cubo['pressure_weather'].astype(float)
cubo['time_hour_weahter'] = pd.to_datetime(cubo['time_hour_weather'], errors='coerce')
```

##### ***a. Cantidad de filas y columnas del cubo***

```python
num_rows, num_cols = cubo.shape
print(f'{num_rows} filas ,{num_cols} columnas')
```

##### ***b. Segmentaci√≥n de columnas por tipo de variable***

```python
continuas = []
discretas = []
categoricas=[]
datetime=[]



for columnas in cubo.columns:
    valores=cubo[columnas]
    if 'hour' in columnas.lower() or 'date' in columnas.lower()or 'month' in columnas.lower()or 'day' in columnas.lower()or 'time' in columnas.lower():
       datetime.append(columnas)
    elif((valores.dtype=='int64')or(valores.dtype=='float64')):
        if(len(valores.unique())<30):
            discretas.append(columnas)                      
        else:
            continuas.append(columnas)
    else:
            categoricas.append(columnas)

print("CONTINUAS")
print(continuas)
print("")
print("DISCRETAS")
print(discretas)
print("")
print("CATEGORICAS")
print(categoricas)
print("")
print("DATE TIME")
print(datetime)
```

##### ***C. Gr√°ficos de variables***

##### ***Discretas***

```python
for columnas in discretas:
    flights_count=cubo[columnas].value_counts()
    x=list(flights_count.index)
    plt.figure(figsize=(4,2))
    plt.bar(x,flights_count)
    plt.title("flights per " + columnas)
    plt.xlabel(columnas)
    plt.ylabel("cantidad")
    print("----")
    plt.show()
```

##### ***Categoricas***

```python
for columnas in categoricas:
    flights_count=cubo[columnas].value_counts()
    x=list(flights_count.index)
    plt.figure(figsize=(4,2))
    plt.bar(x,flights_count)
    plt.title("flights per " + columnas)
    plt.xlabel(columnas)
    plt.ylabel("cantidad")
    print("----")
    plt.show()
```

##### ***Continuas***

```python
for columnas in continuas:
    sns.kdeplot(data=cubo[columnas])
    print("----")
    plt.show()


```

##### ***Fecha y hora***

```python
for columnas in datetime:
    flights_count=cubo[columnas].value_counts().sort_index()
    x=list(flights_count.index)
    plt.figure(figsize=(4,2))
    plt.plot(x,flights_count)
    plt.title("flights per " + columnas)
    plt.xlabel(columnas)
    plt.ylabel("cantidad")
    print("----")
    plt.show()
```
